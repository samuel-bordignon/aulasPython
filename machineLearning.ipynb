```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('datasetnovo.csv')
```

```python
df.head()
```

```python
df.corr()
```

```python
df.info()
```

```python
df.describe()
```

```python
df['horas_estudo_mes'].describe()
```

```python
sns.histplot(data = df, x ='horas_estudo_mes', kde=True)
```

```python
#prepando a variável de entrada
x = np.array(df['horas_estudo_mes'])
```

```python
x
```

```python
type(x)
```

```python
x = x.reshape(-1,1)
```

```python
x
```

```python
y = np.array(df['salario'])
```

```python
plt.scatter(x, y, color = 'blue', label= "Dados Reais Históricos")
plt.xlabel("Horas de Estudo por Mês")
plt.ylabel("Salário")
plt.legend()
plt.show()
```

```python
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
x_teste.shape
```

```python
x_treino.shape
```

```python
y_treino.shape
```

```python
y_teste.shape
```

```python
#cria um modelo de regressão
modelo = LinearRegression()
modelo
```

```python
#treina o modelo
modelo.fit(x_treino, y_treino)
```

```python
# vizualiza a reta de regressõ linear (previsões) e os dados reais usados no treinamento
plt.scatter(x, y, color = 'blue', label= "Dados Reais Históricos")
plt.plot(x, modelo.predict(x), color = 'red', label= "Reta de regressão com as previsões do Modelo")
plt.xlabel("Horas de Estudo por Mês")
plt.ylabel("Salário")
plt.legend()
plt.show()
```

```python
# avalia o modelo nos dados de teste
score = modelo.score(x_teste, y_teste)
print(f"Coeficiente de determinação (R²): {score:.2f}")
```

```python
# intercepto - parâmetro w0
modelo.intercept_
```

```python
# slope - parâmetro w1
modelo.coef_
```

```python
# define um novo valor para horas de estudo
horas_estudo_novo = np.array([[0]])

# faz previsão com o modelo reinado
salario_previsto = modelo.predict(horas_estudo_novo)

print(salario_previsto)
```
